```python
import numpy as np
import matplotlib.pyplot as plt

import mne
from mne.datasets.sleep_physionet.age import fetch_data
from mne.time_frequency import tfr_multitaper

# Update the paths with the actual locations of your files
alice_files = [
    r'sleep-edf-database-expanded-1.0.0/sleep-cassette/SC4001E0-PSG.edf',  # EEG data file
    r'sleep-edf-database-expanded-1.0.0/sleep-cassette/SC4001EC-Hypnogram.edf'  # Annotations file
]

# Load raw EEG data and annotations
raw_train = mne.io.read_raw_edf(alice_files[0], preload=True)
annot_train = mne.read_annotations(alice_files[1])

# Set annotations for the raw data
raw_train.set_annotations(annot_train, emit_warning=False)

# Define the channel types mapping (if needed)
mapping = {'EEG Fpz-Cz': 'eeg', 'EEG Pz-Oz': 'eeg'}
raw_train.set_channel_types(mapping)

# Plot the PSD directly using the built-in method
raw_train.plot_psd(fmin=0.5, fmax=50, average=True)

# If you want to perform time-frequency analysis
# Assuming `epochs` is already created
# tfr = tfr_multitaper(epochs, fmin=0.5, fmax=50, n_cycles=2, time_bandwidth=4.0)
# tfr.plot()
```

```python
ALICE, BOB = 0, 1

[alice_files, bob_files] = fetch_data(subjects=[ALICE, BOB], recording=[1])

mapping = {'EOG horizontal': 'eog',
           'Resp oro-nasal': 'misc',
           'EMG submental': 'misc',
           'Temp rectal': 'misc',
           'Event marker': 'misc'}
```

```python
raw_train = mne.io.read_raw_edf(alice_files[0])
annot_train = mne.read_annotations(alice_files[1])

raw_train.set_annotations(annot_train, emit_warning=False)
raw_train.set_channel_types(mapping)

# plot some data
raw_train.plot(duration=60, scalings='auto')
```

```python
annotation_desc_2_event_id = {'Sleep stage W': 1,
                              'Sleep stage 1': 2,
                              'Sleep stage 2': 3,
                              'Sleep stage 3': 4,
                              'Sleep stage 4': 4,
                              'Sleep stage R': 5}

events_train, _ = mne.events_from_annotations(
    raw_train, event_id=annotation_desc_2_event_id, chunk_duration=30.)
```

```python
# create a new event_id that unifies stages 3 and 4
event_id = {'Sleep stage W': 1,
            'Sleep stage 1': 2,
            'Sleep stage 2': 3,
            'Sleep stage 3/4': 4,
            'Sleep stage R': 5}
```

```python
# plot events
mne.viz.plot_events(events_train, event_id=event_id,
                    sfreq=raw_train.info['sfreq'])

# keep the color-code for further plotting
stage_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
tmax = 30. - 1. / raw_train.info['sfreq']  # tmax in included

epochs_train = mne.Epochs(raw=raw_train, events=events_train,
                          event_id=event_id, tmin=0., tmax=tmax, baseline=None)

print(epochs_train)
```
